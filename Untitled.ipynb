```python
# ==========================================
# 1. Setup Environment
# ==========================================
!pip install transformers[torch] datasets scikit-learn torch accelerate sentencepiece -q

import pandas as pd
import numpy as np
import torch
from torch import nn
from datasets import load_dataset
from transformers import AutoTokenizer, AutoModelForSequenceClassification, Trainer, TrainingArguments
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score

# ==========================================
# 2. Data Loading & Mapping
# ==========================================
print("📥 Loading Dataset...")
raw_dataset = load_dataset("OdiaGenAI/sentiment_analysis_hindi")

# 0: Negative, 1: Neutral, 2: Positive
label_to_id = {"neg": 0, "neu": 1, "pos": 2}

def map_labels(example):
    example["label"] = label_to_id[example["label"]]
    return example

dataset = raw_dataset['train'].map(map_labels)
dataset = dataset.shuffle(seed=42).select(range(1500))
dataset = dataset.train_test_split(test_size=0.2)

# ==========================================
# 3. Model & Tokenization Setup
# ==========================================
model_name = "l3cube-pune/hindi-bert-v2"
tokenizer = AutoTokenizer.from_pretrained(model_name)

def tokenize_function(examples):
    return tokenizer(examples["text"], padding="max_length", truncation=True, max_length=128)

print("🔧 Tokenizing data...")
tokenized_datasets = dataset.map(tokenize_function, batched=True)

device = "cuda" if torch.cuda.is_available() else "cpu"
model = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=3).to(device)

# ==========================================
# 4. Custom Weighted Trainer (The "Bias Fix")
# ==========================================
# We give 'Negative' a weight of 2.0 to force the model to learn it better
weights = torch.tensor([2.0, 1.5, 1.0]).to(device)

class WeightedTrainer(Trainer):
    def compute_loss(self, model, inputs, return_outputs=False, **kwargs):
        labels = inputs.get("labels")
        outputs = model(**inputs)
        logits = outputs.get("logits")
        loss_fct = nn.CrossEntropyLoss(weight=weights)
        loss = loss_fct(logits.view(-1, self.model.config.num_labels), labels.view(-1))
        return (loss, outputs) if return_outputs else loss

# ==========================================
# 5. Metrics & Training Config
# ==========================================
def compute_metrics(eval_pred):
    logits, labels = eval_pred
    predictions = np.argmax(logits, axis=-1)
    # zero_division=0 makes the output clean for the report
    return {
        "accuracy": accuracy_score(labels, predictions),
        "f1": f1_score(labels, predictions, average="weighted"),
        "precision": precision_score(labels, predictions, average="weighted", zero_division=0)
    }

training_args = TrainingArguments(
    output_dir="./results",
    eval_strategy="epoch",        # Corrected for 2026 versions
    save_strategy="epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=16,
    num_train_epochs=3,           # Increased to 3 for better convergence
    weight_decay=0.01,
    logging_steps=10,
    load_best_model_at_end=True,
    metric_for_best_model="f1",
    report_to="none"
)

trainer = WeightedTrainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_datasets["train"],
    eval_dataset=tokenized_datasets["test"],
    compute_metrics=compute_metrics,
)

# ==========================================
# 6. Execution: Train & Evaluate
# ==========================================
print("🚀 Starting Fine-Tuning with Custom Class Weights...")
trainer.train()

print("\n📊 Final Evaluation Results:")
eval_results = trainer.evaluate()
print(eval_results)

# ==========================================
# 7. Inference Test
# ==========================================
def predict_sentiment(text):
    inputs = tokenizer(text, return_tensors="pt", truncation=True, padding=True, max_length=128).to(device)
    with torch.no_grad():
        outputs = model(**inputs)

    probs = torch.nn.functional.softmax(outputs.logits, dim=-1)
    pred_idx = torch.argmax(probs, dim=-1).item()

    labels = ["Negative 👎", "Neutral 😐", "Positive 👍"]
    return labels[pred_idx], probs[0]

test_texts = [
    "यह वीडियो बहुत प्रेरणादायक है!",
    "मुझे आपकी बातें बिल्कुल पसंद नहीं आईं, यह बहुत बुरा था।",
    "ठीक ठाक है, पर जानकारी कम थी।"
]

print("\n🧪 Live Testing (After Bias Correction):")
print("-" * 30)
for text in test_texts:
    label, scores = predict_sentiment(text)
    print(f"Text: {text}")
    print(f"Result: {label}")
    print(f"Confidence - Neg: {scores[0]:.1%}, Neu: {scores[1]:.1%}, Pos: {scores[2]:.1%}")
    print("-" * 30)

# ==========================================
# 8. Save Model
# ==========================================
model.save_pretrained("./josh_hindi_sentiment_model")
tokenizer.save_pretrained("./josh_hindi_sentiment_model")
print("💾 Model saved successfully!")
```

```python
import pandas as pd

# Creating a sample dataset of Hindi comments
data = {
    'comment': [
        "यह कहानी बहुत प्रेरणादायक थी, धन्यवाद जोश टॉक्स!",
        "मुझे यह वीडियो बिल्कुल पसंद नहीं आया, समय की बर्बादी है।",
        "जानकारी अच्छी थी लेकिन आवाज साफ नहीं थी।",
        "क्या आप और भी ऐसे वीडियो बना सकते हैं?",
        "बहुत ही घटिया वीडियो है, कुछ समझ नहीं आया।"
    ]
}

df_sample = pd.DataFrame(data)
df_sample.to_csv('josh_talks_comments.csv', index=False)
print("✅ Sample file 'josh_talks_comments.csv' created!")
```

```python
def analyze_feedback_file(file_path):
    df = pd.read_csv(file_path)
    print(f"📂 Processing {len(df)} comments from {file_path}...")

    results = []
    confidences = []

    for comment in df['comment']:
        sentiment, scores = predict_sentiment(str(comment))
        results.append(sentiment)
        # Store the confidence of the predicted class
        confidences.append(f"{torch.max(scores).item():.2%}")

    df['predicted_sentiment'] = results
    df['confidence'] = confidences

    output_file = "analyzed_feedback.csv"
    df.to_csv(output_file, index=False)
    print(f"✅ Analysis complete! Results saved to '{output_file}'")
    return df

# Run the analysis
final_df = analyze_feedback_file('josh_talks_comments.csv')

# Display the results directly in Colab
final_df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix

# Get predictions for the test set
preds = trainer.predict(tokenized_datasets["test"])
pred_labels = np.argmax(preds.predictions, axis=-1)
true_labels = preds.label_ids

# Create Matrix
cm = confusion_matrix(true_labels, pred_labels)
plt.figure(figsize=(8,6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=["Neg", "Neu", "Pos"],
            yticklabels=["Neg", "Neu", "Pos"])
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Sentiment Confusion Matrix')
plt.show()
```
